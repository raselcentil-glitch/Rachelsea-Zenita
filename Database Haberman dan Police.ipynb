code haberman


```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('haberman.csv')

# Tampilkan data awal
print(df.head())

# Rename kolom biar jelas
df.columns = ['Age', 'Year', 'Nodes', 'Survival']

# Info dataset
print(df.info())

# Statistik deskriptif
print(df.describe())

# Cek missing values
print(df.isnull().sum())

# Distribusi target (Survival)
sns.countplot(x='Survival', data=df)
plt.title('Distribusi Survival')
plt.show()

# Histogram setiap fitur
df.hist(figsize=(10,8))
plt.show()

# Boxplot untuk melihat outlier
sns.boxplot(data=df)
plt.show()

# Korelasi
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Pairplot
sns.pairplot(df, hue='Survival')
plt.show()
```

code police

```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('police.csv')

# Tampilkan data awal
print(df.head())

# Info dataset
print(df.info())

# Statistik deskriptif
print(df.describe())

# Cek missing values
print(df.isnull().sum())

# Distribusi data kategorikal (contoh kolom 'violation' jika ada)
if 'violation' in df.columns:
    sns.countplot(y='violation', data=df)
    plt.title('Jenis Pelanggaran')
    plt.show()

# Distribusi numerik
df.hist(figsize=(12,10))
plt.show()

# Korelasi (hanya numerik)
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()

# Contoh analisis berdasarkan gender (jika ada)
if 'driver_gender' in df.columns:
    sns.countplot(x='driver_gender', data=df)
    plt.title('Distribusi Gender')
    plt.show()

# Contoh hubungan pelanggaran vs gender
if 'violation' in df.columns and 'driver_gender' in df.columns:
    sns.countplot(x='driver_gender', hue='violation', data=df)
    plt.show()
```
